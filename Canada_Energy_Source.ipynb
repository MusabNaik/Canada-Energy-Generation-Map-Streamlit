```python
import pandas as pd
df = pd.read_csv('./25100015.csv')
```

```python
df
```

```python
df.dtypes
```

```python
df['year'] = df['REF_DATE'].str[:4]
```

```python
df['REF_DATE'] = pd.to_datetime(df['REF_DATE'], format='%Y-%m')
```

```python
df2 = df[(df['REF_DATE'] >= '2016-01') &
         (df['REF_DATE'] <= '2022-01') &
         (df['GEO'] != 'Canada') &
   (df['Class of electricity producer'] == 'Total all classes of electricity producer') &
   (~df['Type of electricity generation'].isin(['Conventional steam turbine',
                                                'Internal combustion turbine',
                                                'Combustion turbine',
                                                'Total electricity production from non-renewable combustible fuels',
                                                'Total electricity production from biomass',
                                                'Total all types of electricity generation']) )]
```

```python
def prov_code(row):
  if row['GEO'] == 'Alberta':
    return 'AB'
  elif row['GEO'] == 'British Columbia':
    return 'BC'
  elif row['GEO'] == 'Manitoba':
    return 'MB'
  elif row['GEO'] == 'New Brunswick':
    return 'NB'
  elif row['GEO'] == 'Newfoundland and Labrador':
    return 'NL'
  elif row['GEO'] == 'Northwest Territories':
    return 'NT'
  elif row['GEO'] == 'Nova Scotia':
    return 'NS'
  elif row['GEO'] == 'Nunavut':
    return 'NU'
  elif row['GEO'] == 'Ontario':
    return 'ON'
  elif row['GEO'] == 'Prince Edward Island':
    return 'PE'
  elif row['GEO'] == 'Quebec':
    return 'QC'
  elif row['GEO'] == 'Saskatchewan':
    return 'SK'
  elif row['GEO'] == 'Yukon':
    return 'YT'
  else:
    return row['GEO']
```

```python
df2['Prov'] = df2.apply(prov_code,axis=1)
```
